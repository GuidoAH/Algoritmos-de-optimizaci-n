#1.Método descenso gradiente (parte grupal)
Función a estudiar $f(x,y) = x^2 + y^2$

```python
#importar librerias
import math
import matplotlib.pyplot as plt
import numpy as np
import random

#definir variables y parámetros
gamma = 0.05
maxit = 500
tol = 0.05
rango = 3
xp = random.uniform(-rango, rango)
yp = random.uniform(-rango, rango)
x0 = np.array([xp, yp])
plt.plot(x0[0],x0[1], "o", c="white")

#definir funciones
df = lambda x: [2*x[0], 2*x[1]]
f = lambda x: x[0]**2 + x[1]**2

# #preparar el plot
resolucion = 30
u = np.linspace(-rango,rango,resolucion)
x,y = np.meshgrid(u,u)
z = x**2 + y**2
contour = plt.contourf(x,y,z)
cbar = plt.colorbar(contour)

def desc_grad(df, gamma, tol, maxit, x0):
  i = 0
  while i <= maxit:
    x1 = x0 - gamma * np.array(df(x0))
    if (x1 - x0).all() < tol:
      return (f"Solución: punto{x0}, f(x,y)={f(x0)}")
    x0 = x1
    plt.plot(x0[0],x0[1], "o", c="red")
    i += 1
    if i == maxit:
      plt.plot(x0[0],x0[1], "o", c="yellow")
      return (f"Solución: punto:{x0}, f(x,y)={f(x0)}")

desc_grad(df, gamma, tol, maxit, x0)
```

#2. Método descenso gradiente (parte extra | parte individual)

Función a estudiar $f(x,y) = sin(\frac{1}{2}x^2 - \frac{1}{4}y^2+3)cos(2x+1-e^y)$

```python
#importar librerias
import math
import matplotlib.pyplot as plt
import numpy as np
import random

#definir variables y parámetros
gamma = 0.05
maxit = 100
tol = 0.05
rango = 5
#definimos punto inicial
xp = random.uniform(-rango, rango)
yp = random.uniform(-rango, rango)
x0 = np.array([xp, yp])
plt.plot(x0[0],x0[1], "o", c="white")

#definir funciones
f = lambda X: np.sin(1/2 * X[0]**2 - 1/4 * X[1]**2 + 3) * np.cos(2 * X[0] + 1- np.e**X[1])
df = lambda x: [2*x[0], 2*x[1]]

def df(point):
    h = 0.01
    grad = np.zeros(2)
    for it in range(len(point)):
        T = np.copy(point)
        T[it] = T[it] + h
        grad[it] = (f(T) - f(point)) / h
    return grad

#preparar el plot
resolucion = 30
u = np.linspace(-rango,rango,resolucion)
x,y = np.meshgrid(u,u)
z = np.sin(1/2 * x**2 - 1/4 * y**2 + 3) * np.cos(2 * x + 1- np.e**y)
contour = plt.contourf(x,y,z)
cbar = plt.colorbar(contour)

def desc_grad(df, gamma, tol, maxit, x0):
  i = 0
  while i <= maxit:
    grad = df(x0)
    x1 = x0 - gamma * np.array(df(x0))
    if (x1 - x0).all() < tol:
    # if (x1[0] - x0[0]) < tol and (x1[1] - x0[1]) < tol:
      return (f"Solución: punto{x0}, f(x,y)={f(x0)}")
    x0 = x1
    plt.plot(x0[0],x0[1], "o", c="red")
    i += 1
    if i == maxit:
      plt.plot(x0[0],x0[1], "o", c="yellow")
      return (f"Solución: punto:{x0}, f(x,y)={f(x0)}")

desc_grad(df, gamma, tol, maxit, x0)
```

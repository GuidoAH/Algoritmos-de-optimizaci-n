```python
#importar librerias
import math
import matplotlib.pyplot as plt
import numpy as np
import random

#definir variables y par치metros
gamma = 0.05
maxit = 500
tol = 0.05
rango = 3
xp = random.uniform(-rango, rango)
yp = random.uniform(-rango, rango)
x0 = np.array([xp, yp])
plt.plot(x0[0],x0[1], "o", c="white")

#definir funciones
df = lambda x: [2*x[0], 2*x[1]]
f = lambda x: x[0]**2 + x[1]**2

# #preparar el plot
resolucion = 30
u = np.linspace(-rango,rango,resolucion)
x,y = np.meshgrid(u,u)
z = x**2 + y**2
contour = plt.contourf(x,y,z)
cbar = plt.colorbar(contour)

def desc_grad(df, gamma, tol, maxit, x0):
  i = 0
  while i <= maxit:
    x1 = x0 - gamma * np.array(df(x0))
    if (x1 - x0).all() < tol:
      return (f"Soluci칩n: punto{x0}, f(x,y)={f(x0)}")
    x0 = x1
    plt.plot(x0[0],x0[1], "o", c="red")
    i += 1
    if i == maxit:
      plt.plot(x0[0],x0[1], "o", c="yellow")
      return (f"Soluci칩n: punto:{x0}, f(x,y)={f(x0)}")

desc_grad(df, gamma, tol, maxit, x0)
```
